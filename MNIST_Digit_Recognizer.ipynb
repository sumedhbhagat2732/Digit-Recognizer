### Preparing Training and Testing Dataset


```python
# Split the data into features and target variable
X = df.drop('label', axis=1)
y = df['label']
```

```python
# Scale the pixel values to have a range of 0 to 1
X = X / 255.0
```

```python
#Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Reset Index
X_train.reset_index(inplace=True)
X_test.reset_index(inplace=True)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
#Drop the index column
X_train.drop('index',axis=1,inplace=True)
X_test.drop('index',axis=1,inplace=True)
```

```python
print('Number of training samples :',len(y_train))
print('Number of testing samples :',len(y_test))
```

There are 3664 training and 917 testing samples after the split

### Building Classification Models and making Predictions 

1) The 6 Classification models which have been considered are KNeighborsClassifier, Naive Bayes Classifier, Support Vector Classifier, Logistic Regression, Decision Tree Classifier and Random Forest Classifier.

2) 5 steps have been implemented to make predictions and check the accuracy score:-

    a) Initializing object for the classifier model.
    b) Fitting the model on training data.
    c) Predicting the results on test data using the same object created in step 1.
    d) Calculating the accuracy score 
    e) Making a Classification report for the results.

#### KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

knn = KNeighborsClassifier() 
knn.fit(X_train, y_train) 
y_pred_knn = knn.predict(X_test) 
acc_knn = accuracy_score(y_test, y_pred_knn)
report_knn = classification_report(y_test,y_pred_knn)
print('Accuracy Score :',acc_knn)
print(report_knn)
```

#### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB() 
gnb.fit(X_train, y_train) 
y_pred_gnb = gnb.predict(X_test)
acc_gnb = accuracy_score(y_test, y_pred_gnb)
report_gnb = classification_report(y_test,y_pred_gnb)
print('Accuracy Score :',acc_gnb)
print(report_gnb)
```

#### Support Vector Classifier

```python
from sklearn.svm import SVC

svc = SVC(random_state=0) 
svc.fit(X_train, y_train) 
y_pred_svc = svc.predict(X_test) 
acc_svc = accuracy_score(y_test, y_pred_svc)
report_svc = classification_report(y_test,y_pred_svc)
print('Accuracy Score :',acc_svc)
print(report_svc) 
```

####Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='saga', max_iter=10000,random_state=0) 
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
acc_lr = accuracy_score(y_test, y_pred_lr)
report_lr = classification_report(y_test,y_pred_lr)
print('Accuracy Score :',acc_lr)
print(report_lr) 
```

#### Decision Trees Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(random_state=0) 
dtc.fit(X_train, y_train)
y_pred_dtc = dtc.predict(X_test)
acc_dtc = accuracy_score(y_test, y_pred_dtc)
report_dtc = classification_report(y_test,y_pred_dtc) 
print('Accuracy Score :',acc_dtc)
print(report_dtc)
 
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=0) 
rfc.fit(X_train, y_train) 
y_pred_rfc = rfc.predict(X_test)
acc_rfc = accuracy_score(y_test, y_pred_rfc)
report_rfc = classification_report(y_test,y_pred_rfc) 
print('Acuuracy Score :',acc_rfc)
print(report_rfc)
```

#### Compare Accuracies of Different Classification Models in one Plot


```python
# Create a bar plot
models = ['KNN','Naive Bayes', 'SVC','Logistic Regression','Decision Tree', 'Random Forest']
accuracies = [acc_knn,acc_gnb, acc_svc, acc_lr,acc_dtc, acc_rfc ]
fig, ax = plt.subplots(figsize = (15, 5))
ax.bar(models, accuracies)

# Add data labels to each bar
for i, v in enumerate(accuracies):
    ax.text(i, v, str(round(v, 2)), ha='center', va='bottom')

# Set axis labels and title
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of Diiferent Classification Models')

plt.show()
```

### Hyper Parameter Tuning

1) After the comparison , it has been found that the top 3 models in terms of accuracy are Support Vector Classifier, Random Forest Classifier and KNeighbors Classifier. 

2) In the next step we will try to do Hyper Parameter tuning on these three models and see if the accuracy can be increased.

3) Grid SearchCV will be used to find optimal hyperparameters for each model and find the best accuracy using them.

```python
# Define the hyperparameters to tune for each model
from sklearn.model_selection import GridSearchCV
rf_param_grid = {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10, 15]}
svm_param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
knn_param_grid = {'n_neighbors': [3, 5, 7, 9], 'weights': ['uniform', 'distance']}

# Create a list of tuples containing each model and its corresponding hyperparameter grid
models = [(rfc, rf_param_grid), (svc, svm_param_grid), (knn, knn_param_grid)]

# Loop through each model and its hyperparameter grid, and fit a GridSearchCV object to the training data
for model, param_grid in models:
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    
    # Print the best hyperparameters and accuracy score for each model
    print(model.__class__.__name__)
    print("Best hyperparameters: ", grid_search.best_params_)
    print("Best accuracy: ", grid_search.best_score_)
    
```

We can see that after hyper parameter tuning , not much improvement has been seen in the accuracies of the models and thus we will choose the top model i.e. Support Vector Classifier to do prediction of our original testing dataset.
